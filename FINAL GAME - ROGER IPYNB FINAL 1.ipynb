<h3> The Story </h3>

Tom and Jerry is one of the most popular cartoonish shows of all time with a 9.0/10 rating on IMDB. The story is quite simple, there are two main characters, Tom, the cat and Jerry, the mouse. Tom is always chasing Jerry and they set each other traps, but surprisingly, Jerry always gets away and wins. 

The characters: 
-	Tom: The cat. He is always chasing Jerry and setting up traps. 
-	Jerry: The mouse. In the show, he is pictured as the main character as well as the good one, while Tom is considered the bad one. 

<h3> The Game </h3>

This game has been inspired by the 1940–1968 TV show, but giving it a little twist, where Tom is the main character and Jerry is not. 

The objective of the game is for the user, in this case being Tom, to chase down and eat Jerry, while setting up traps. 

The game has 4 phases:

     1. Phase 1 (Intro) 
     2. Phase 2 (Find Tom) 
     3. Phase 3 (Find Cheese)
